```python
import pandas as pd
import numpy as np
```

## Name of the columns:

```python
columns = ['age','sex','chest_ache','precision','cholesterol',
           'sugar_high','electro','max_beats','angina','depression_st',
           'slope','vessels','thal','diagnosis']
```

```python
data = pd.read_csv('processed.cleveland.csv', names=columns ,na_values='?')
```

```python
data
```

### values 1,2,3 & 4 in diagnosis means there's a illness

```python
data.loc[:, 'diagnosis'] = data.diagnosis.apply(lambda x: int(x > 0))
```

```python
data
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(12, input_dim=13, activation='relu'))#entrada/oculta layer
model.add(Dense(8, activation='relu'))#oculta layer
model.add(Dense(1, activation='sigmoid'))#salida layer
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
X = data.drop('diagnosis',axis=1)
Y = data['diagnosis']

print(X.shape, Y.shape)
```

```python
X.isnull().sum()
```

```python
X = X.fillna(-1)
X.isnull().sum()
```

```python
model.summary()
```

```python
model.fit(X, Y, validation_split= 0.1, epochs=150, batch_size=10)
```
